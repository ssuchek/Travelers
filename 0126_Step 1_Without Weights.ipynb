```python
missed_items_all = data[no_match_mask].groupby(by='item_description').sum()['counter'].sort_values(ascending=False)

missed_items_lf = data[lf_mask & no_match_mask].groupby(by='item_description').sum()['counter'].sort_values(ascending=False)

missed_items_sf = data[sf_mask & no_match_mask].groupby(by='item_description').sum()['counter'].sort_values(ascending=False)

missed_items_sq = data[sq_mask & no_match_mask].groupby(by='item_description').sum()['counter'].sort_values(ascending=False)

missed_items_ea = data[ea_mask & no_match_mask].groupby(by='item_description').sum()['counter'].sort_values(ascending=False)
```
